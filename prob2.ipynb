```python
pd.read_sql("""SELECT title, edition, copyright FROM titles 
    WHERE copyright > '2016'""", connection)
```

```python
pd.read_sql("""SELECT id, first, last FROM authors WHERE last 
    LIKE 'D%'""", connection, index_col=['id'])
```

```python
pd.read_sql("""SELECT id, first, last FROM authors WHERE first 
    LIKE '_b%'""", connection, index_col=['id'])
```

```python
pd.read_sql('SELECT title FROM titles ORDER BY title ASC', connection)
```

```python
pd.read_sql("""SELECT id, first, last FROM authors ORDER BY 
    last, first""", connection, index_col=['id'])
```

```python
pd.read_sql("""SELECT id, first, last FROM authors ORDER BY 
    last DESC, first ASC""", connection, index_col=['id'])
```

```python
pd.read_sql("""SELECT isbn, title, edition, copyright FROM titles 
    WHERE title LIKE '%How to Program' ORDER BY title""", connection)
```

```python
pd.read_sql("""SELECT first, last, isbn FROM authors INNER JOIN 
    author_ISBN ON authors.id = author_ISBN.id ORDER BY 
    last, first""", connection).head()
```
